# 01. Series
1D data(정수, 실수, 문자열 등)

## Series 객체 생성
ex) 1월부터 4월까지 평균 온도 데이터(-20,-10,10,20)

```python
# Visual Python: Logic > code
temp = pd.Series([-20,-10,10,20]) # index 자동으로 생성됨
print(temp)
temp
```

```python
# Visual Python: Logic > code
temp[0] #1월 온도
temp[2] #3월 온도
```

```python
# Visual Python: Logic > comment
# 존재하지 않은 index에 접근 시 에러
# temp['June']
```

## Series 객체 생성(Index 지정)

```python
# Visual Python: Logic > code
temp = pd.Series([-20,-10,10,20], index = ['Jan','Feb','Mar','Apr'])
temp
```

```python
# Visual Python: Logic > code
temp['Apr'] # index Apr (4월)에 해당하는 데이터 출력
```

```python
# Visual Python: Logic > comment
# 존재하지 않은 index에 접근 시 에러
# temp['June']
```

# 05. 데이터 확인

```python
# Visual Python: Logic > code
import pandas as pd

data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index = ['1번','2번','3번','4번','5번','6번','7번','8번'])
df.index.name = '지원번호'

df.to_csv('score.csv', encoding='utf-8-sig', index = False)  # index 는 빼고 저장
df.to_csv('score.txt', sep='\t')  # tab으로 구분된 txt파일
df.to_excel('score.xlsx')
```

# 07. 데이터 선택(loc)
**이름**을 이용하여 원하는 row에서 원하는 col 선택

---

loc에서 슬라이싱 할 시, : 뒤의 데이터 모두 포함함 주의

```python
df.loc['1번']
```

```python
# Visual Python: Data Analysis > Subset
df.loc[(df.index == '1번'), :]
```

```python
# Visual Python: Logic > code
df.loc[['1번','2번'],['영어','수학']]   # index 1번,2번에 해당하는 영어, 수학 데이터
df.loc['1번':'5번', '국어':'사회']  # 1번부터 4번 아니고 5번까지임!! 포함함!!
```

# 08. 데이터 선택(iloc)
**i**nteger **loc**ation

*   위치(정수)를 이용하여 row에서 원하는 col 선택



```python
# Visual Python: Data Analysis > Subset
df.iloc[0, :]
```

```python
# Visual Python: Data Analysis > Subset
df.iloc[0:5, :]    # 0~4번째 위치(행)의 데이터
## 5번째 데이터 포함 안 함 주의
```

```python
# Visual Python: Data Analysis > Subset
df.iloc[4, 2]    # 5번 학생의 키 데이터 (4번째 위치(행)의 2번째 열(키) 데이터)
```

```python
# Visual Python: Data Analysis > Subset
df.iloc[[0,1], 2]   #0,1 번째 위치의 학생의 두 번째(키) 데이터
```

```python
# Visual Python: Data Analysis > Subset
df.iloc[[0,1], [2,5]]  #0,1 번째 위치의 학생의 2,5 번째(키, 수학) 데이터
```

```python
# Visual Python: Data Analysis > Subset
df.iloc[0:5, 3:8]    # 0~4 번째 위치의 학생 중 3~7번째 데이터 (국어:사회)
```
